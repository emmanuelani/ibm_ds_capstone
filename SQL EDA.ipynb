## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
For Example "Landing_Outcome"**

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
cur.execute("SELECT DISTINCT Launch_Site FROM ..spacex")

results = cur.fetchall()

for row in results:
    print(row[0])
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
cur.execute("SELECT TOP 5 * FROM ..spacex WHERE Launch_Site LIKE 'CCA%'")

results_2 = cur.fetchall()

for row in results_2:
    print(row[:])
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# cur.execute('SELECT customer, SUM(PAYLOAD_MASS_KG) FROM spacex GROUP BY customer')
cur.execute("SELECT customer, SUM(PAYLOAD_MASS_KG)\
        FROM spacex\
        WHERE customer = 'NASA (CRS)'\
        GROUP BY customer")

results_3 = cur.fetchall()

for row in results_3:
    print(row[:])
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
cur.execute("SELECT Booster_Version, AVG(PAYLOAD_MASS_KG)\
    FROM spacex\
    WHERE Booster_Version = 'F9 v1.1'\
    GROUP BY Booster_Version")

results_4 = cur.fetchall()

for row in results_4:
    print(row[:])
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
cur.execute("SELECT TOP 1 Date\
    FROM spacex\
    WHERE Landing_Outcome = 'Success (ground pad)'\
    ORDER BY Date")

results_5 = cur.fetchall()

for row in results_5:
    print(row[0])

```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
cur.execute("SELECT Booster_Version\
    FROM spacex\
    WHERE Landing_Outcome = 'Success (drone ship)'\
    AND PAYLOAD_MASS_KG BETWEEN 4000 AND 6000")

results_6 = cur.fetchall()

for row in results_6:
    print(row[0])
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
# Successful mission
cur.execute("SELECT COUNT(Mission_Outcome) FROM spacex\
    WHERE Mission_Outcome LIKE '%Success%'")

results_7 = cur.fetchall()

for row in results_7:
    print(f"Successful missions: {row[0]}")

# Failed missions
cur.execute("SELECT COUNT(Mission_Outcome) FROM spacex\
    WHERE Mission_Outcome LIKE '%Failure%'")

results_7 = cur.fetchall()

for row in results_7:
    print(f"Failed Missions: {row[0]}")
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
cur.execute("SELECT Booster_Version, PAYLOAD_MASS_KG \
    FROM spacex \
    WHERE PAYLOAD_MASS_KG = (SELECT MAX(PAYLOAD_MASS_KG) \
                            FROM spacex) \
    ORDER BY Booster_Version")

results_8 = cur.fetchall()

for result in results_8:
    print(result)
```

### Task 9

##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
cur.execute("SELECT Date, Booster_Version, Launch_Site, Landing_Outcome\
            FROM spacex\
            WHERE Date BETWEEN '2014-12-31' AND '2016-01-01'\
            AND Landing_Outcome='Failure (drone ship)'")

results_9 = cur.fetchall()

for row in results_9:
    print(row)
```

### Task 10

##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
cur.execute("SELECT Landing_Outcome, COUNT(Landing_Outcome)\
    FROM spacex\
    WHERE date BETWEEN '2010-06-04' AND '2017-03-20'\
    GROUP BY Landing_Outcome\
    ORDER BY COUNT(Landing_Outcome) DESC")

results_10 = cur.fetchall()

for result in results_10:
    print(result)
```
